# MNIST Handwritten Digit Recognition

## Feed-Forward Neural Network Using TensorFlow Keras

This notebook is designed for third-year students.

### Steps

1. Load the MNIST dataset
2. Preprocess the images
3. Design the neural-network layers
4. Compile the model
5. Evaluate before training
6. Train using gradient descent
7. Evaluate after training
8. Plot accuracy and loss
9. Display predictions and a confusion matrix

# 1. Import Libraries

````python
# Uncomment only when TensorFlow is not installed.
# !pip install tensorflow matplotlib

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

print("TensorFlow version:", tf.__version__)
````

# 2. Load the MNIST Dataset

````python
(x_train, y_train), (x_test, y_test) = (
    tf.keras.datasets.mnist.load_data()
)

print("Training images:", x_train.shape)
print("Training labels:", y_train.shape)
print("Test images:", x_test.shape)
print("Test labels:", y_test.shape)
````

## Display Sample Images

````python
plt.figure(figsize=(10, 4))

for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i], cmap="gray")
    plt.title(f"Label: {y_train[i]}")
    plt.axis("off")

plt.tight_layout()
plt.show()
````

# 3. Preprocess the Images

Pixel values range from `0` to `255`.

Dividing by `255` converts them into values between `0` and `1`.

````python
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

print("Minimum pixel value:", x_train.min())
print("Maximum pixel value:", x_train.max())
````

# 4. Design the Feed-Forward Neural Network

```text
28 × 28 image
      ↓
Flatten layer: 784 values
      ↓
Dense layer: 128 neurons, ReLU
      ↓
Dense layer: 64 neurons, ReLU
      ↓
Output layer: 10 neurons, Softmax
```

````python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(28, 28)),

    tf.keras.layers.Flatten(),

    tf.keras.layers.Dense(
        128,
        activation="relu"
    ),

    tf.keras.layers.Dense(
        64,
        activation="relu"
    ),

    tf.keras.layers.Dense(
        10,
        activation="softmax"
    )
])

model.summary()
````

# 5. Compile the Model

- **Optimizer:** SGD
- **Loss:** Sparse categorical cross-entropy
- **Metric:** Accuracy

SGD stands for Stochastic Gradient Descent.

````python
optimizer = tf.keras.optimizers.SGD(
    learning_rate=0.1
)

model.compile(
    optimizer=optimizer,
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

print("Model compiled successfully.")
````

# 6. Evaluate Before Gradient-Descent Training

At this point, the model has random weights.

`model.fit()` has not yet been called, so the model has not learned from the dataset.

````python
initial_loss, initial_accuracy = model.evaluate(
    x_test,
    y_test,
    verbose=0
)

print("Before training")
print("Test loss:", initial_loss)
print("Test accuracy:", initial_accuracy)
````

# 7. Train the Model Using Gradient Descent

During training, TensorFlow performs:

1. Forward propagation
2. Loss calculation
3. Backpropagation
4. Gradient-descent weight update

````python
history = model.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)
````

# 8. Evaluate the Trained Model

````python
final_loss, final_accuracy = model.evaluate(
    x_test,
    y_test,
    verbose=0
)

print("After training")
print("Test loss:", final_loss)
print("Test accuracy:", final_accuracy)
````

## Compare Accuracy Before and After Training

````python
print(f"Accuracy before training: {initial_accuracy:.4f}")
print(f"Accuracy after training:  {final_accuracy:.4f}")

accuracy_improvement = final_accuracy - initial_accuracy

print(f"Accuracy improvement:     {accuracy_improvement:.4f}")
print(
    f"Percentage-point improvement: "
    f"{accuracy_improvement * 100:.2f}"
)
````

````python
stages = ["Before Training", "After Training"]
accuracy_values = [initial_accuracy, final_accuracy]

plt.figure(figsize=(6, 4))
plt.bar(stages, accuracy_values)
plt.ylim(0, 1)
plt.ylabel("Test Accuracy")
plt.title("Accuracy Before and After Gradient Descent")

for i, value in enumerate(accuracy_values):
    plt.text(i, value + 0.02, f"{value:.3f}", ha="center")

plt.show()
````

# 9. Plot Training and Validation Accuracy

````python
plt.figure(figsize=(8, 5))

plt.plot(
    history.history["accuracy"],
    label="Training Accuracy"
)

plt.plot(
    history.history["val_accuracy"],
    label="Validation Accuracy"
)

plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training and Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()
````

# 10. Plot Training and Validation Loss

````python
plt.figure(figsize=(8, 5))

plt.plot(
    history.history["loss"],
    label="Training Loss"
)

plt.plot(
    history.history["val_loss"],
    label="Validation Loss"
)

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.legend()
plt.grid(True)
plt.show()
````

# 11. Make Predictions

````python
probabilities = model.predict(
    x_test,
    verbose=0
)

predicted_labels = np.argmax(
    probabilities,
    axis=1
)

print("Actual labels:   ", y_test[:10])
print("Predicted labels:", predicted_labels[:10])
````

## Display Predicted Digits

````python
plt.figure(figsize=(12, 6))

for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_test[i], cmap="gray")

    predicted = predicted_labels[i]
    actual = y_test[i]
    confidence = probabilities[i][predicted]

    plt.title(
        f"Actual: {actual}\n"
        f"Predicted: {predicted}\n"
        f"Confidence: {confidence:.2f}"
    )

    plt.axis("off")

plt.tight_layout()
plt.show()
````

# 12. Confusion Matrix

````python
confusion = tf.math.confusion_matrix(
    y_test,
    predicted_labels,
    num_classes=10
).numpy()

print(confusion)
````

````python
plt.figure(figsize=(8, 7))
plt.imshow(confusion, cmap="Blues")
plt.title("MNIST Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.colorbar()

ticks = np.arange(10)
plt.xticks(ticks, ticks)
plt.yticks(ticks, ticks)

for row in range(10):
    for column in range(10):
        plt.text(
            column,
            row,
            confusion[row, column],
            ha="center",
            va="center"
        )

plt.tight_layout()
plt.show()
````
